**A. Ромб**

Вычислите площадь ромба со диагоналями d1 и d2

```python
def rhombus(d1, d2):
    return d1*d2/2

test(rhombus(1, 0), 0)
test(rhombus(435, 752), 163560)
test(rhombus(547.426845, 33.346728), 9127.44705005)
```

**B. Налог**

Вычислите налог с зарплаты по прогрессивной шкале с тремя ступенями по правилу:

&nbsp; • доход меньше \\$10k — 10%    
&nbsp; • доход между \\$10k и \\$20k — 20% с превышения над \\$10k  
&nbsp; • доход свыше \\$20k — 30% с превышения над \\$20k

```python
def tax(x):
    if(x <= 10e3):
        return x*0.1
    if(10e3 < x < 20e3):
        return 10e3*0.1 + (x-1e3)*0.2
    return 10e3*0.1  + (x-10e3 - x%10e3)*0.2 + (x - 20e3)*0.3

test(tax(100), 10.)
test(tax(1000), 100.)
test(tax(10000), 1000.)
test(tax(10100), 1020.)
test(tax(20100), 3030.)
```

 **C. Ветро-холодовой индекс**
 
Мера «жесткости» холода, учитывающая температуру и скорость ветра (в Австралии также влажность).

При сильном ветре температура кажется ниже, чем есть на самом деле, и опасность обморожения (frostbite) выше,
чем без ветра:

Формула (более точная, но слегка не совпадает с рис. выше) такая:

```python
def wind_chill(t, v):
    return 13.12 + 0.6215*t - 11.37*v**0.16 + 0.3965*t*v**0.16

test(wind_chill(-24, 5), -28.81631)
test(wind_chill(-20, 12), -28.032769)
test(wind_chill(-19, 20), -29.217109)
```

#### D. Ряд
Для заданного k найдите сумму ряда с относительной погрешностью менее $10^{-6}$

$$\sum\limits_{n=0}^\infty \frac{(-1)^n}{(kn)!}$$

```python
def series(k):
    n = 0
    previous = 0
    #current = (-1)**n / factorial(k*n)
    current = 1
    while current == 0 or abs((current - previous)/current) > 1e-6:
        previous = current 
        n += 1
        current += (-1)**n / (factorial(k*n))
     
    return current

test(series(1), 1/e)
test(series(2), cos(1))
test(series(3), 0.834719)

```
