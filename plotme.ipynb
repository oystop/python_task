### Задание 

Реализуйте функцию plotme(f, a, b), которая бы строила график математической функции, заданной функцией $f(x)$ на промежутке от $a$ до $b$, 500 точек. 

В тех же осях постройте производную (взятую численным дифференцированием любым способом).


```python
import matplotlib.pyplot as plt
import numpy as np
from array import *
import math
```

```python
def func(x):
    return x**2
```

```python
def derivative(f, x):
    delta = 1e-5
    df = (f(x+delta) - f(x-delta)) / (2*delta)
    return df
```

```python
def plotme(f,a,b):
    
    x_vals = np.linspace(a,b, 500)
    y_vals = f(x_vals)
    dev_y_vals = derivative(f, x_vals)
    
    plt.plot(x_vals,y_vals, 'b', label = 'график функции')
    plt.plot(x_vals, dev_y_vals, 'm', label = 'график производной')
    plt.grid()
    plt.title('График функции и ее производной')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.show()
    
```

```python
plotme(func, -1, 3)
```

```python
def my_sin(x):
    return np.sin(x)
```

```python
plotme(my_sin, -2*math.pi, 2*math.pi)
```

```python
def my_log(x):
    return np.log(x)
```

```python
plotme(my_log, 1,5)
```
